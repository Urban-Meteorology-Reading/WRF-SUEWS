```python
df=pd.read_csv('./runs/run_Colombo/Input/Cl_2016_data_60_not_corrected.txt',sep=' ')

df.head()
```

```python
def get_time(row):
    year=int(row['iy'])
    day=int(row['id'])
    hour=int(row['it'])
    minute=int(row['imin'])
    return datetime.datetime(year, 1, 1) + datetime.timedelta(day - 1,hours=hour,minutes=minute)
```

```python
temp=df.apply(get_time,axis=1)+pd.Timedelta(5.5,'h')
df['iy']=[x.year for x in temp]
df['id']=[x.timetuple().tm_yday for x in temp]
df['it']=[x.hour for x in temp]
df['imin']=[x.minute for x in temp]
```

```python
df.to_csv('./runs/run_Colombo/Input/Cl_2016_data_60.txt',sep=' ',index=0)
```
